FROM THIS IT SEEMS THAT THE BEST IS THE `power_plants_output_hydro` GENERATES
THE MOST ENERGY AND IS THE MOST STABLE 

GENEAL FUNCTION TO DETRMINE THAT

```python
def renewable_highest_mean(df, renewable_cols):
    renewable = df[renewable_cols]
    names = renewable[renewable_cols].mean().sort_values(ascending=False).index
    values = renewable[renewable_cols].mean().sort_values(ascending=False).values
    return names, values

def renewable_highest_std(df, renewable_cols):
    renewable = df[renewable_cols]
    names = renewable[renewable_cols].std().sort_values(ascending=True).index
    values = renewable[renewable_cols].std().sort_values(ascending=True).values
    return names, values
```

```python
renewable_highest_mean(df, renewable_cols)
```

```python
renewable_highest_std(df, renewable_cols)
```
